```python
import pandas as pd
import numpy as np
import folium
```

```python
path="C:/data/environment/" #각자 파일 경로

data=pd.read_excel(path + 'location with pm201903.xlsx')
```

```python
from folium import plugins

location=np.array([data['lat'],data['lng']]).T
m = folium.Map(location=[37.612, 127.03], zoom_start=11)

popups=[str(i) for i in range(len(data.index))]

plugins.MarkerCluster(location, popups=popups).add_to(m)
m
```

```python
map = folium.Map(location=[37.612, 127.03], zoom_start=12)

for n in data.index:
    folium.CircleMarker([data['lat'][n], data['lng'][n]], radius=data['pm10_201903'][n]*0.25, 
                        color='#3186cc', fill_color='#3186cc').add_to(map)
map
```
